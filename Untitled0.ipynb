```python
#La aguja de Buffon 
#importamos el módulo que vamos a usar 
import random   
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl
import math
d=0
z=150

puntosx1=[]
puntosy1=[]
for i in range(z):
  puntosx1.append(random.uniform(0,5))
  puntosy1.append(random.uniform(0,5))


fig = plt.figure(1, figsize=(5, 5)) 
for i in range(z):
  a=math.sqrt((puntosx1[i]*puntosx1[i])+(puntosy1[i]*puntosy1[i]))
  b1=puntosx1[i]
  c1=puntosy1[i]
  plt.scatter(b1, c1, s=5, c="red")
  plt.scatter(((b1/a)+b1), ((c1/a)+c1), s=5, c="red")
  x = np.linspace(b1,(b1/a)+b1, 100)
  y = (c1/b1)*x
  plt.scatter(x, y, s=5, c="red")
  for j in range(-5,6,1):
    if j>b1 and j<(b1+(b1/a)):
      d=d+1
pl.xlim(-1, 6)
pl.ylim(-1, 6)
for i in range(-5, 6,1):
  plt.axvline(x=i, ymin=0, ymax=5)   
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
print((2*z)/d)
```
